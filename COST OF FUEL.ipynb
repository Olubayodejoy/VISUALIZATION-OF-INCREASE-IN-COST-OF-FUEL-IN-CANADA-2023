```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Month': ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December'],
        'Price': [150, 155, 160, 165, 170, 175, 180, 185, 190, 185, 180, 175]}

fuel_df = pd.DataFrame(data)

```

```python
fuel_df
```

```python
plt.figure(figsize=(10,6))
plt.plot(fuel_df['Month'], fuel_df['Price'], marker='o')
plt.title('Monthly Petrol Prices in Canada (2023)')
plt.xlabel('Month')
plt.ylabel('Price (cents/litre)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
plt.xticks(rotation=75)
sns.barplot(x='Month', y='Price', data=fuel_df)
plt.show()
```

```python

plt.figure(figsize=(8,8))
plt.pie(fuel_df['Price'], labels=fuel_df['Month'], autopct='%1.1f%%', startangle=140)

plt.title('Petrol Price Increases in Canada (2023)', fontsize=14)

plt.show()
```
